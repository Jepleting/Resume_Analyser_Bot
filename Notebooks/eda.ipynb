```python
#plot the distribution
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,6))
sns.countplot(data=df, y='Category', order=df['Category'].value_counts().index)
plt.title("Job Role Distribution")
plt.xlabel("Count")
plt.ylabel("Job Role")
plt.tight_layout()
plt.show()
```

```python
#word count per resume
df['Word_Count'] = df['Resume'].apply(lambda x: len(str(x).split()))
df['Word_Count'].hist(bins=20)
plt.title("Resume Word Count Distribution")
plt.xlabel("Words")
plt.ylabel("Frequency")
plt.show()
```

```python
from wordcloud import WordCloud 
```

```python
#wordcloud
text_blob = ' '.join(df['Resume'].astype(str))
wordcloud = WordCloud(width=1000, height=500, background_color='white').generate(text_blob)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.title("WordCloud - All Resumes")
plt.show()
```
